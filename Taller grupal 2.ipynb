```python
import numpy as np
import skimage 
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import random
from skimage import io
from skimage import feature
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
#Punto 1

tamaño = 300 #Establecemos el tamaño de la imagen (300x300)
imagen = [] #Iniciamos una lista en donde se encontrará nuestra imagen

for i in range(tamaño): #Por cada fila se inicia una lista
  fila = []
  for j in range(tamaño): #Por cada elemento de esa fila... 
    if j >= i:             #... si su índice de columna es mayor
      rojo = 255             # que su indice de fila..
      verde = 255         #.. el pixel correspondiente será blanco
      azul = 255
    else:                 
      rojo = 0             #En caso contrario será negro
      verde = 0         
      azul = 0
    pixel = [rojo, verde, azul]    #Establecemos el pixel
    fila.append(pixel)    #Cada elemento de la fila tendrá un pixel
  imagen.append(fila)       #Añadimos las filas a la imagen final.

plt.imshow(imagen)  #Mostramos la imagen
```

```python
#Suavizar y resaltar contornos
imagen = skimage.data.coffee() #con la misma data de la imgaen de una taza de caje la colocamos como "imagen"

imagen2 = rgb2gray(imagen) #se denomina a imagen2 como la imagen en escala de grises gracias al codigo rgb2gray, mostrando solo tonos blancos y negros

bordes = feature.canny(imagen2, sigma = 2) #se demonima bordes a la funcion de feature.canny que se encarga de denotar los bordes gracias al gradiente que es
# la transcicion en un rango de colores, gracias al comando de imagen2 se denota mas facil al trancision de negro a blanco, el valor sigma hace referencia a la "sensibilidad" del filtro

plt.imshow(bordes) #se muestra la imagen ya con los bordes resaltados gracias a plt.imshow()

plt.show() # se muestra una escala de medida en los bordes
```

```python
# Eliminacion del ruido
from skimage.filters.rank import median #Se importa desde los datos de filtros de skimage el filtro de median 
ejemplo = skimage.data.skin() #importamos desde los datos de skimage la imagen de la piel humana
grises = rgb2gray(ejemplo) #definimos "grises" a la variable la cual se le aplica el codigo rgb2gray() el cual coloca la imagen en un tono de
# escala de grises
acabado = median(grises) #definimos "acabado" a la varibale la cual aplicamos la linea de codigo median() este filtro lo que hace es
# eliminar el ruido de una imagen(pixeles repetidos a travez de toda la imagen que la perturba) eliminandolo y dejando la imagen
# en un tono verde pero sin los pixeles de ruido
plt.imshow(acabado) #mostramos la imagen a travez de plt.imshow ya acabada
plt.figure()
```
